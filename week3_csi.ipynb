```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

print(titanic.head())

plt.figure(figsize=(6, 4))
sns.countplot(x='survived', data=titanic, palette='Set2')
plt.title('Survival Count')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x='sex', hue='survived', data=titanic, palette='Set1')
plt.title('Survival by Gender')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=titanic, palette='pastel')
plt.title('Passenger Class Distribution')
plt.show()

plt.figure(figsize=(8, 4))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.show()

sns.catplot(x="class", hue="survived", col="sex", kind="count", data=titanic, palette='muted')
plt.suptitle('Survival by Class and Gender', y=1.03)
plt.show()
```
